```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv("C://Users//vishesh//Downloads//Prices.csv")
reg = linear_model.LinearRegression()
reg.fit(data[['price']],data.area)
%matplotlib inline
plt.xlabel('Price')
plt.ylabel('Area (in sq. m)')
plt.scatter(data[['price']],data.area)
plt.plot(data.price,reg.predict(data[['price']]),color='green')
```
